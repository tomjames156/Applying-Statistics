```python
# A visualization that proves that the normal distribution is an approximation of the probability binomial distribution

## The data is obtained by flipping a coin 10 times and then moving left (heads) or right (tails)
## The probability binomial distribution is calculated by assuming that left is the successful outcome, right is unsuccessful and the number of trials is 10
## The final position is evaluated by taking left as negative and right as positive
## The normal distribution is obtained by the formula using the mean and standard deviation and taking the x to be the number of successful outcomes

### The dataset is computed using a spreadsheet software
```

```python
import pandas as pd

df = pd.read_excel("normal_binomial_compared.xlsx", skiprows=2)
df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns

df.plot(kind='line', x='Final Position', y=['Normal Distribution', 'Binomial Distribution'])
plt.title("Normal Distribution vs Binomial Distribution")
plt.ylabel("Probability")
plt.show()
```
