```python
# A line graph that shows the convergence between the probabilities obtained using the normal distribution and binomial probability distributions
# It proves that the normal distribution is a good estimate of the binomial distribution and that the probabilities get closer as the mean number of trials increases
# The data involves the random steps taken which require flipping a coin to go left(heads) or right(tails) to get the same final position
# A left step is negative and a right step is positive 
# The number of steps taken are multiples of 10 i.e 10, 20, 30, e.t.c

# The dataset was computed using a spreadsheet
```

```python
import pandas as pd

df = pd.read_excel('convergence.xlsx', skiprows=2)
df1 = df.head(17)
df1
```

```python
from matplotlib import pyplot as plt

df1[['Difference', "Mean"]].plot(kind='line', x='Mean')
plt.title("Converge of Normal and Binomial Probability Distributions")
plt.ylabel("Probability Difference")
plt.xlabel("Mean Number of Trials")
plt.show()
```

```python
df2 = df.tail(17)
df2
```
