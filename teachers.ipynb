```python
# Drawing a bar graph that shows a group 20 teachers and their favourite courses

import pandas as pd

df = pd.DataFrame({
    'courses': ['Language', 'History', 'Geometry', 'Physics', 'Chemistry'],
    'favourites': [7, 3, 9, 0, 1]
})

df
```

```python
from matplotlib import pyplot as plt
```

```python
df.plot(kind='bar', x='courses', y='favourites')
plt.xticks(rotation=45)
plt.ylabel('Number of teachers')
plt.title('Favourite Courses')
plt.legend(loc='upper right', title='KEY')
plt.xlabel('Course')
plt.show()
```
