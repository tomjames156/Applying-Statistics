```python
# Residual plot example with Seaborn

import pandas as pd

df = pd.DataFrame({
    "x": [1, 2, 2, 3],
    "y": [1, 2, 3, 6]
})

df
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

fig, ax = plt.subplots()
plt.title("Residual Plot")
sns.residplot(data=df, x='x', y='y', ax=ax)
ax.set_xlabel('X')
ax.set_ylabel('Residuals')
plt.show()
```
