```python
import numpy as np
import pandas as pd

# A dataframe for a population of 100 people from age 10 to 25
population_ages = np.random.randint(10, 26, size=100)
population_df = pd.DataFrame({'age': population_ages})
population_df
```

```python
# Getting biased variance and std deviation for the population

stnd_dev = float(population_df.age.std(ddof=0))
stnd_dev = round(stnd_dev, 2)
variance = float(population_df.age.var(ddof=0))
variance = round(variance, 2)
pop_mean = float(population_df.age.mean())
pop_mean = round(pop_mean, 1)

pop_mean, stnd_dev, variance
```

```python
import seaborn as sns
from matplotlib import pyplot as plt

# Plot the population data
sns.histplot(data=population_df, x='age', discrete=True)
plt.title(f"Age Distribution of Customers\nMean: {pop_mean}    Std Dev: {stnd_dev}    Variance: {variance}")
plt.ylabel("No of customers")
plt.xlabel('Ages')
plt.show()
```

```python
def get_sample_variance(sample_size):
    """Returns a mean sample variance for a specific sample size"""

    sample_variances = []

    # Get 100 samples of the sample size
    for i in range(100):
        sample = population_df.sample(sample_size)
        # Get the variance of each sample 
        sample_variance = float(sample.age.var(ddof=0))
        sample_variances.append(sample_variance)

    sample_variances = np.asarray(sample_variances)
    # Get the mean variance of the 100 samples
    mean_variance = sample_variances.mean()
    # Get the ratio of the mean variance to the population variance
    sample_population_var_ratio = mean_variance / variance
    return round(sample_population_var_ratio, 2)
```

```python
variance_df = pd.DataFrame({'sample size': [2, 3, 4, 5, 6, 7, 8, 9, 10]})
variance_df
```

```python
variance_df['biased_variance'] = variance_df['sample size'].apply(lambda x: get_sample_variance(x))
variance_df
```

```python
sns.barplot(data=variance_df, x='sample size', y='biased_variance')
plt.ylabel("Biased Sample variance / Population Variance")
plt.xlabel('Sample Sizes (number of customers)')
plt.title("Sample Variance vs Sample Size")
plt.show()
```

```python
def get_sample_mean_and_variance(sample_size):
    """Returns the mean and variance for a sample taken from a population"""

    sample = population_df.sample(sample_size)
    sample_mean = sample.mean().item() # converts it to float
    sample_var = sample.var(ddof=0).item()
    return round(sample_mean, 2), round(sample_var, 2)
```

```python
from random import randint

sample_sizes = []
sample_means = []
sample_variances = []

# Get mean and variances for 100 random sample sizes
for i in range(1000):
    sample_size = randint(2, 12) # sample sizes from 2 to 12\
    mean_customers, biased_variance = get_sample_mean_and_variance(sample_size)
    # Add the size, mean and variance to the corresponding list
    sample_sizes.append(sample_size)
    sample_means.append(mean_customers)
    sample_variances.append(biased_variance)
```

```python
# Create a dataframe from the data collected
random_samples_mean_variance = pd.DataFrame({
    "sample size": sample_sizes,
    "sample mean": sample_means,
    "sample variance": sample_variances
})
random_samples_mean_variance
```

```python
sns.scatterplot(data=random_samples_mean_variance, x='sample mean', y='sample variance', hue='sample size')
plt.axvline(x=17.3, linestyle='--')
plt.axhline(y=21.2, linestyle='--')
plt.ylabel("Sample Variance")
plt.xlabel("Sample Mean")
plt.title("Sample Statistics\nLines indicate Population statistics")
plt.show()
```

```python
# The graphs above illustrate the fact that dividing by n - 1 produces an unbiased variance
# It also shows that smaller sample sizes are most likely to give a wrong estimate of the mean and variance
```
