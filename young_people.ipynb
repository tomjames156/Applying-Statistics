```python
# Visualizing a data set with a box and whisker plot
import pandas as pd

young_df = pd.DataFrame({'age': [1, 1, 6, 13, 13, 14, 14, 14, 15, 15, 16, 18, 18, 18, 19]})
young_df.shape
```

```python
from matplotlib import pyplot as plt

# young_df.plot(kind='line')
# plt.show()

plt.boxplot(young_df['age'], vert=False)
plt.show()
```

```python
import seaborn as sns

sns.set_style("whitegrid") 
sns.boxplot(data=young_df, x='age')
plt.show()
```

```python
young_df.describe()

import numpy as np

q75, q50, q25 = np.percentile(young_df, [75, 50, 25], interpolation='midpoint')
q75, q50, q25
```
